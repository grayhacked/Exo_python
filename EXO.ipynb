```python
"""
Enoncé : Ou gen yon adrès IP.

Ou dwe rantre nan yon pòt(PORT) ki ouvri. Ou pa gen anpil tan pou skane tout pòt yo, men ou konnen pòt ki ouvri a, li egal total tout dijit yo ki nan adrès IP a, miltiplye pa premye dijit ki nan adrès IP a.

Mande itilizatè a pou l tape adrès IP a, epi afiche pòt ki ouvri a

NB: Tout adrès ip separe ak pwen(.)

Egzanp: Si itilizatè a antre adrès IP
"127.0.0.1" ap bay rezilta
(1+2+7+0+0+1) * 1 = 11 =>
Donk pòt ki ouvri a se 11
"""
```

```python
#fonction principale
def ip_to_sum(ip_address):
    digits = ip_address.split('.')
    valeur = list(digits[0]) + list(digits[1]) + list(digits[2]) + list(digits[3])
    first_digits = int(valeur[0])
    somme = 0
    for chiffre in valeur:
        somme += int(chiffre)
    return somme*first_digits
```

```python
#demande a l'utilisateur de rentrer une adresse IP
IP = input("Enter IP address: ")
output = ip_to_sum(IP)
print("the available port is:", output)
```

```python
def divisible_par_4(n):
    if n % 4 != 0:
        return "OK"
    else:
        return "NOK"
```
